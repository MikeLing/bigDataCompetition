```python
%pylab inline
%matplotlib inline

import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plot

train_data = []
```

```python
def load_data(file_path):
    with open(file_path, 'r') as fin:
        for line in fin:
            data = line.split(' ')
            mouse_id = data[0]
            mouse_move = data[1]
            mouse_target = data[2]
            labels = data[3]
            train_data.append({"id": mouse_id, "movement" : mouse_move, "target" : mouse_target, "label": labels})
```

```python
load_data("data/train_sample.txt")
```

```python
mouse_movement1 = np.asarray([d.split(",") for d in train_data[0]["movement"].split(';')][:-1], dtype=np.float64)
mouse_movement2 = np.asarray([d.split(",") for d in train_data[1]["movement"].split(';')][:-1], dtype=np.float64)
mouse_movement3 = np.asarray([d.split(",") for d in train_data[2]["movement"].split(';')][:-1], dtype=np.float64)
mouse_movement4 = np.asarray([d.split(",") for d in train_data[3]["movement"].split(';')][:-1], dtype=np.float64)
```

```python
fig = plot.figure(figsize=(10,8))
plot.xlabel('X')
plot.ylabel('Y')
plot.title('FOR ID 1')
plot.plot(mouse_movement1[:, 0], mouse_movement1[:, 1])
plot.show()

fig = plot.figure(figsize=(10,8))
plot.xlabel('X')
plot.ylabel('Y')
plot.title('FOR ID 1')
plot.scatter(mouse_movement1[:, 0], mouse_movement1[:, 1])
plot.show()

fig = plot.figure(figsize=(10,8))
plot.xlabel('X')
plot.ylabel('Y')
plot.title('FOR ID 2')
plot.plot(mouse_movement2[:, 0], mouse_movement2[:, 1])
plot.show()

fig = plot.figure(figsize=(10,8))
plot.xlabel('X')
plot.ylabel('Y')
plot.title('FOR ID 2')
plot.scatter(mouse_movement2[:, 0], mouse_movement2[:, 1])
plot.show()

fig = plot.figure(figsize=(10,8))
plot.xlabel('X')
plot.ylabel('Y')
plot.title('FOR ID 3')
plot.plot(mouse_movement3[:, 0], mouse_movement3[:, 1])
plot.show()

fig = plot.figure(figsize=(10,8))
plot.xlabel('X')
plot.ylabel('Y')
plot.title('FOR ID 3')
plot.scatter(mouse_movement3[:, 0], mouse_movement3[:, 1])
plot.show()

fig = plot.figure(figsize=(10,8))
plot.xlabel('X')
plot.ylabel('Y')
plot.title('FOR ID 4')
plot.plot(mouse_movement4[:, 0], mouse_movement4[:, 1])
plot.show()

fig = plot.figure(figsize=(10,8))
plot.xlabel('X')
plot.ylabel('Y')
plot.title('FOR ID 4')
plot.scatter(mouse_movement4[:, 0], mouse_movement4[:, 1])
plot.show()
```
