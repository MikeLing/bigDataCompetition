```python
%pylab inline
%matplotlib inline

import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plot
import mpl_toolkits.mplot3d.axes3d as p3

train_data = []
plotData = []
```

```python
def load_data(file_path):
    with open(file_path, 'r') as fin:
        for line in fin:
            data = line.split(' ')
            mouse_id = data[0]
            mouse_move = data[1]
            mouse_target = data[2]
            labels = data[3]
            train_data.append({"id": mouse_id, "movement" : mouse_move, "target" : mouse_target, "label": labels})
```

```python
r = [a for a in train_data if a['label'] == 0]
```
